# Micro-wave

```python
from scipy.special import jv

Delta_meV = 0.18

V_0 = 0.5
A_0 = 0.5
V_0_mV = V_0 * Delta_meV
A_0_mV = A_0 * Delta_meV

nu_GHz = 10.0

t_0_ps = 1e3 / nu_GHz
t_ps = np.linspace(0, 540, 1001)  # ps
dt_ps = t_ps[1] - t_ps[0]
dt_s = dt_ps * 1e-12

V = V_0 + A_0 * np.cos(2 * np.pi * t_ps / t_0_ps)
V_mV = V * Delta_meV

m = 1
nu_mV = nu_GHz * sc.h_pVs
alpha_0 = m * A_0_mV / nu_mV

# dimensionless (radians)
phi = m * np.cumsum(V_mV) * dt_ps * 2 * np.pi / sc.h_pVs * 1e-3
U = np.exp(-1j * phi)

A = np.array([0.5, 1, 1.5, 2.0])
A_mV = A * Delta_meV
alpha = m * A_mV / nu_mV

n0 = np.linspace(0, 13, 14)
n1 = np.linspace(0, 13, 131)
j0 = jv(n0[None, :], alpha[:, None])
j1 = jv(n1[None, :], alpha[:, None])
```

```python
fig, ax = sc.get_figure(figsize=(2.0, 1.2), padding=(0.3, 0.2))

ax.plot(
    t_ps,
    V,
    color=sc.seeblau100,
)

ax.set_ylim(-0.1, 1.2)

xlabel = "$t$ (ps)"
ylabel = "$eV\\,/\\,\\Delta_0$"
title = "theory/basics/mw-V"

sc.theory_layout(fig, ax, title, xlabel, ylabel)
```

```python
fig, ax = sc.get_figure(figsize=(2.0, 1.2), padding=(0.3, 0.2))

ax.plot(
    t_ps,
    phi,
    color=sc.seeblau100,
)

ax.set_ylim(-5, 90)

xlabel = "$t$ (ps)"
ylabel = "$\\phi_q(t)$"
title = "theory/basics/mw-phi"

sc.theory_layout(fig, ax, title, xlabel, ylabel)
```

```python
fig, ax = sc.get_figure(figsize=(2.0, 1.2), padding=(0.3, 0.2))

ax.plot(
    t_ps,
    np.real(U),
    color=sc.seeblau100,
)

ax.set_ylim(-1.1, 1.35)

xlabel = "$t$ (ps)"
ylabel = "$\\Re\\,\\!(U_q(t))$"
title = "theory/basics/mw-ReU"

sc.theory_layout(fig, ax, title, xlabel, ylabel)
```

```python
fig, ax = sc.get_figure(figsize=(2.0, 1.2), padding=(0.3, 0.2))

color = [
    sc.seeblau100,
    sc.seegrau80,
    sc.seeblau65,
    sc.seegrau65,
    sc.seeblau35,
    sc.seegrau35,
]

for i, a in enumerate(alpha):
    ax.plot(
        n0 * nu_GHz,
        j0[i, :],
        ".",
        ms=2,
        zorder=i + 4,
        color=color[i],
        label=f"{A[i]}",
    )
    ax.plot(
        n1 * nu_GHz,
        j1[i, :],
        "-",
        lw=0.5,
        zorder=i + 4,
        color=color[i],
    )

ax.legend(
    title="$eA\\,/\\,\\Delta_0 = $",
    handlelength=0.1,
    loc="upper right",
    bbox_to_anchor=(1.05, 1.15),
    frameon=False,
    fontsize=6,
    title_fontsize=6,
    ncols=2,
    columnspacing=1.0,
)
ax.set_ylim(-0.4, 0.8)

xlabel = "$\\nu$ (GHz)"
ylabel = "$J_n(\\alpha_q)$"
title = "theory/basics/mw-Jn"

sc.theory_layout(fig, ax, title, xlabel, ylabel)
```
